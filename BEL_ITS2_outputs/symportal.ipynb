```r
#cclean to just include Condition columns and colony'
colony_clean=colony_immune[c('colony', 'X062019_Condition', 'X052022_Condition', 'X122022_Condition', 'X092023_Condition', 'X112023_Condition', 'X122023_Condition','X012024_Condition','X022024_Condition', 'X042024_Condition', 'X062024_Condition', 'X082024_Condition')]
head(colony_clean)
```

```r
#rename condition columns to corresponding Monthyear
colony_clean <- colony_clean %>%
  rename_with(
    .cols = matches("^X\\d{6}_Condition$"),
    .fn = ~ str_extract(., "\\d{6}")
  )
```

```r
head(colony_clean)
```

```r
#pivot longer colony to join all date columns into one monthyear column
colony_long <- colony_clean %>%
  pivot_longer(
    cols = matches("^\\d{6}$"),      # after renaming
    names_to = "MonthYear",
    values_to = "Condition"
  )
head(colony_long)
```

```r
#make monthyear column in meta from tubelabel_species
#cleaning samdt
meta <- meta[meta$Date_ITS2 != 'NA', ]
# Create a new data frame with the species, date, and transect columns
# Example samdt (make sure your data is a data.table)
setDT(meta)
# Split Tubelabel_species by "_" and assign new columns
meta[, c("date_sampled") := {
  parts <- strsplit(Tubelabel_species, "_")
  list(
    sapply(parts, `[`, 1)  # date is first element
  )
}]

#changing format of date_sampled to be Jan 2024
meta$date_sampled <- as.factor(meta$date_sampled)
# Convert to Date format 
meta$DateFormatted <- as.Date(paste0(meta$date_sampled, "01"), format = "%m%Y%d")
# now will have a new column that has Month Year instead of date_sampled
meta$MonthYear <- format(meta$DateFormatted, "%b %Y")
meta$MonthYear <- factor(meta$MonthYear, levels = unique(meta$MonthYear))
```

```r
meta_clean <- meta[, c("V1", "Date_Extracted", "Raw_ng_ul", "Extraction_physical_location", "Date_Enriched","Microbe_Location", "Date_16S", "16S double band", "date_sampled", "transect", "species", "V13", "DateFormatted") := NULL]
```

# use updated meta sheet to add date_its2 and health status to symportal sheet

```r
symportal <- read.csv("/home/caroline_desouza_uml_edu/run3_SymPortal_datasheet.csv",
                      header = FALSE, stringsAsFactors = FALSE)
pcr=read.csv(file="/home/caroline_desouza_uml_edu/meta_PCR.csv", header=TRUE, row.names=1)
```

```r
head(symportal)
head(pcr)
```

```r
# Make row 2 the header
colnames(symportal) <- symportal[1, ]
# Delete rows 1 and 2
symportal <- symportal[-c(1), ]
```

```r
head(symportal)
```

```r
symportal$sample_name <- gsub("-", "_", symportal$sample_name)
head(symportal)
```

```r
#get rid of empty rows
symportal <- symportal[!is.na(symportal$sample_name), ]

names(symportal) <- make.names(names(symportal), unique = TRUE)
#reanme samplenames to tubelabel species
symportal <- symportal %>%
  rename(Tubelabel_species = sample_name)


#merge new colony_long and meta
symportal_update <- merge(symportal, pcr,
                     by = c("Tubelabel_species"),
                     all = FALSE)
```

```r
head(symportal_update)
```

## put in file names

```r
symportal_update$Tubelabel_species <- gsub("_", "-", symportal_update$Tubelabel_species)
head(symportal_update)
```

```r
pathF="/scratch/workspace/caroline_desouza_uml_edu-microbe-Run3/Run3_trim/ITS2/R1"
pathR="/scratch/workspace/caroline_desouza_uml_edu-microbe-Run3/Run3_trim/ITS2/R2"

# Convert the character vector into a list
r1 <- as.data.frame(list.files(pathF))

r2 <- as.data.frame(list.files(pathR))
head(r1)
```

```r
fnFs = sort(list.files(pathF, pattern="_R1_ITS2.fastq", full.names = TRUE))
fnRs = sort(list.files(pathR, pattern="_R2_ITS2.fastq", full.names = TRUE))
r1$Tubelabel_species = sapply(strsplit(basename(fnFs), "_"), `[`, 1)
r2$Tubelabel_species = sapply(strsplit(basename(fnRs), "_"), `[`, 1)
head(r1)
```

```r
# fill in this symportal_update$fastq_fwd_file_name  by $Tubelabel_species
#merge new colony_long and meta
symportal_update <- merge(symportal_update, r1,
                     by = c("Tubelabel_species"),
                     all = FALSE)

# fill in this symportal_update$fastq_fwd_file_name  by $Tubelabel_species
#merge new colony_long and meta
symportal_update <- merge(symportal_update, r2,
                     by = c("Tubelabel_species"),
                     all = FALSE)
head(symportal_update)
```

```r
write.csv(symportal_update, file="/home/caroline_desouza_uml_edu/run3_symportal_update.csv")
```
